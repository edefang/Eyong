## Data cleaning
Clean the data by handling missing values and outliers.


Impute missing values, handle outliers in 'salary', 'salary_in_usd', and 'experience' columns, standardize categories in 'job_title', 'company_size', and 'remote_ratio', and remove duplicate rows.



```python
import pandas as pd
import numpy as np

# Check available columns
print(df.columns)

# Missing Value Imputation (no changes needed here)
for col in df.columns:
    if df[col].isnull().any():
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = df[col].fillna(df[col].median())
        else:
            df[col] = df[col].fillna(df[col].mode()[0])

df.dropna(inplace=True)

# Outlier Treatment (correct column name and convert to numeric)
def cap_outliers(series, percentile=95):
    upper_limit = np.percentile(series, percentile)
    return np.clip(series, a_min=series.min(), a_max=upper_limit)

for col in ['salary', 'salary_in_usd']:
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col] = cap_outliers(df[col])

# Handle outliers in 'experience_level' (convert to numerical representation)
experience_mapping = {'EN': 1, 'MI': 2, 'SE': 3, 'EX': 4}
df['experience_level'] = df['experience_level'].map(experience_mapping)
df['experience_level'] = cap_outliers(df['experience_level'])

# Categorical Variable Consistency (no changes needed here)
job_title_mapping = {
    title: title.lower().replace('sr.', 'senior').replace('jr.', 'junior')
    for title in df['job_title'].unique()
}
df['job_title'] = df['job_title'].map(job_title_mapping)
df['company_size'] = df['company_size'].str.lower()
df['remote_ratio'] = df['remote_ratio'].astype(str)


# Duplicate Removal (no changes needed here)
df.drop_duplicates(inplace=True)

display(df.head())
```

## Data wrangling
Prepare the data for analysis and modeling.


```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Define columns for one-hot encoding
categorical_cols = ['work_year', 'experience_level', 'employment_type', 'job_title', 'salary_currency', 'employee_residence', 'remote_ratio', 'company_location', 'company_size']

# Create the column transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols)
    ],
    remainder='passthrough'
)

# Apply one-hot encoding
df_encoded = pd.DataFrame(preprocessor.fit_transform(df))

# Get feature names - corrected to get the correct number of features
feature_names = list(preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols))
numerical_cols = ['salary', 'salary_in_usd']
feature_names.extend(numerical_cols)

# Ensure the number of column names matches the number of columns in the DataFrame
if len(feature_names) != df_encoded.shape[1]:
    print(f"Warning: Number of feature names ({len(feature_names)}) does not match number of columns in DataFrame ({df_encoded.shape[1]}).")
    # Adjust the feature names to match the number of columns
    feature_names = feature_names[:df_encoded.shape[1]]

df_encoded.columns = feature_names

# Scale numerical features
scaler = MinMaxScaler()
df_encoded[numerical_cols] = scaler.fit_transform(df_encoded[numerical_cols])

display(df_encoded.head())
```

## Data analysis
Analyze the relationship between salary and other variables.
Calculate average salary by job title, correlation between salary and numerical features, and analyze the relationship between salary and categorical features.



Calculate average salary by job title, correlation between salary and numerical features, and analyze the relationship between salary and categorical features.



```python
# Identify the one-hot encoded job title columns
job_title_cols = [col for col in df_encoded.columns if col.startswith('job_title_')]

# Calculate average salary for each job title (using one-hot encoded columns)
average_salary_by_job_title = {}
for col in job_title_cols:
  average_salary_by_job_title[col] = df_encoded[df_encoded[col] == 1]['salary'].mean()
average_salary_by_job_title_series = pd.Series(average_salary_by_job_title).sort_values(ascending=False)
print("Average Salary by Job Title (Descending Order):\n", average_salary_by_job_title_series)

# Calculate correlation between salary and other numerical features (excluding one-hot encoded job titles)
numerical_features = df_encoded.select_dtypes(include=['number']).columns
numerical_features = numerical_features.drop(['salary_in_usd'])
numerical_features = [col for col in numerical_features if not col.startswith('job_title_')]
correlation_matrix = df_encoded[numerical_features].corr()
print("\nCorrelation Matrix:\n", correlation_matrix['salary'])

# Analyze relationship between salary and categorical features (using original columns from df, as they haven't been scaled)
categorical_features = ['experience_level', 'company_size', 'remote_ratio']
for feature in categorical_features:
    average_salary_by_category = df.groupby(feature)['salary'].mean().sort_values(ascending=False)
    print(f"\nAverage Salary by {feature}:\n", average_salary_by_category)
```
