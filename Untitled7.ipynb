## Data cleaning

### Subtask:
Clean the data by handling missing values and outliers in the 'avg_occupancy_proportion' column.


**Reasoning**:
Impute missing values in 'avg_occupancy_proportion' using the median, handle outliers using winsorization, and check for and remove duplicates.



```python
import numpy as np

# Impute missing values in 'avg_occupancy_proportion' with the median
median_occupancy = df['avg_occupancy_proportion'].median()
df['avg_occupancy_proportion'] = df['avg_occupancy_proportion'].fillna(median_occupancy)

# Winsorize outliers in 'avg_occupancy_proportion'
from scipy.stats.mstats import winsorize
df['avg_occupancy_proportion'] = winsorize(df['avg_occupancy_proportion'], limits=[0.05, 0.05])

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Data Consistency Check (ensure correct data types)
# Check data types and make any necessary conversions
display(df.info())
display(df.head())
```

## Data wrangling

### Subtask:
Prepare the data for visualization and analysis by creating a 'Year' column (if not already present) and categorizing unemployment rates.


**Reasoning**:
Create the 'Year' column if it doesn't exist, categorize the unemployment rate, and verify the changes.



```python
# Check if 'Year' column exists; if not, create it
if 'Year' not in df.columns:
    df['Year'] = df['year']

# Categorize unemployment rate
# Define thresholds using quantiles
low_threshold = df['unemployment_rate'].quantile(0.33)
high_threshold = df['unemployment_rate'].quantile(0.66)

def categorize_unemployment(rate):
    if rate <= low_threshold:
        return 'Low'
    elif rate <= high_threshold:
        return 'Medium'
    else:
        return 'High'

df['Unemployment_Category'] = df['unemployment_rate'].apply(categorize_unemployment)

# Ensure correct data types
df['Year'] = df['Year'].astype(int)
df['Unemployment_Category'] = df['Unemployment_Category'].astype('category')

# Display first few rows and data type info
display(df.head())
display(df.info())
```

## Data visualization

### Subtask:
Generate a bar chart showing average office leasing rates across different regions.


**Reasoning**:
Calculate the average leasing rates for each region and generate a bar chart.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average leasing rates per region
average_leasing_rates = df.groupby('market')['leasedSF'].mean()

# Create bar chart
plt.figure(figsize=(10, 6))  # Adjust figure size for better readability
sns.barplot(x=average_leasing_rates.index, y=average_leasing_rates.values, palette='viridis')
plt.title('Average Office Leasing Rates Across Regions')
plt.xlabel('Region')
plt.ylabel('Average Leasing Rate (leasedSF)')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability

# Add data labels on top of each bar
for i, v in enumerate(average_leasing_rates.values):
    plt.text(i, v, f'{v:.0f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
